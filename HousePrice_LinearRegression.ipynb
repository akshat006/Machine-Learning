```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
pd.options.display.max_rows = 4000
pd.options.display.max_columns = 4000
from scipy import stats as st
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('E:/Netzwerk/Practice/HousePrices.csv')
```

```python
df.head()
```

```python
df['SaleCondition'].unique()
```

```python
df.isnull().sum()
```

```python
na_number_per_col = df.isnull().sum()
```

```python
# Remove Columns having more 50% null values

remove_set = set()

for i in na_number_per_col.index:
    if na_number_per_col[i] > 600:
        remove_set.add(i)
```

```python
df.drop(remove_set, axis=1, inplace = True)
```

```python
df.head()
```

```python
df.drop('Id', inplace=True, axis = 1)
```

```python
#Fill Null values
na_number_per_col = df.isnull().sum()
for i in na_number_per_col.index:
    if na_number_per_col[i] > 0 and df[i].dtypes != 'O':
        #print(i)
        mean = df[i].mean()
        df[i].fillna(mean, inplace=True)
    else:
        df[i].fillna(method='bfill', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
numerical_data = df.select_dtypes(include=[np.number]).columns
categorical_data = df.select_dtypes(exclude=[np.number]).columns
```

```python
label_encoder = preprocessing.LabelEncoder()

for i in categorical_data:
    df[i] = label_encoder.fit_transform(df[i])
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
```

```python
X.head()
```

```python
y = df.iloc[:,-1]
```

```python
y.head()
```

```python
# Performing Normalization

X = preprocessing.StandardScaler().fit_transform(X)
X
```

```python
# Diving into test and train data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
ypredict = regressor.predict(X_test)
```

```python
r2_score(y_test, ypredict)
```

```python
prediction_output = pd.DataFrame(data=[y_test.values, ypredict], index = ['y_test','ypredict']).transpose()
```
